# More Indexing

Last time we practiced some indexing, but let's try some more indexing to get some better practice. You will find indexing to be one of the most useful aspects of working with dataframes, and you will likely use this for any research project involving data science and anaysis.

There are a few ways to index using pandas, and we saw a couple last time. Here, we will use a pandas attribute called `loc` that can be used on any dataframe.

You can use this attribute to look at only specific columns, as follows:

`df.loc[:,["Column A", "Column B"]]`

You can also filter your results. For example, if you want to see pressure for every day that both rained and had a temperature above 10 degrees Celcius, you could do the following:

`df.loc[(df["Precip Type"] == "rain") & (df["Temperature (C)"]>20.0), ["Precip Type", "Temperature (C)", "Pressure (millibars)"]]`

In the above code, the first argument is `(df["Precip Type"] == "rain") & (df["Temperature (C)"]>20.0)` and it describes the boolean conditions that need to be met for the row to be included in the returned dataframe.

The second argument is `["Precip Type", "Temperature (C)", "Pressure (millibars)"]` and it names the columns to be included in the returned dataframe (and the order of these columns)

Let's run it and see what it does!



```
pressure_filtered = df.loc[(df["Precip Type"] == "rain") & (df["Temperature (C)"]>20.0), \
                           ["Precip Type", "Temperature (C)", "Pressure (millibars)"]]
display(pressure_filtered)
```

You'll notice that, although there are only 20391 rows, the index label of the last row is 96452. This is because this index label corresponds to the its row in the original dataframe. Thus, you can use this index label to match rows from this filtered dataframe to the original weather dataframe.

# Grouping data

These measures of central tendency and spread are inherently limited in what they can tell us, because they operate across all data points.

For example, it seems intuitive that the average temperature may be strongly linked to the precipation type. For that reason, we may want average temperature of only when it rains and average temperature of only when it snows.

We can run a test to see whether these two things may actually be linked in the way we suspect.

You can do this using the `groupby()` function on the dataframe. As the argument to this function, pass in the the columns you want to group by in an array. This function doesn't return exactly a dataframe, but you can still use the central tendecy and spread measure functions on it. You do not need to specify an axis argument.

For example, below we get the `Summary` and `Humidity` columns using the `loc` method of indexing. Then we group by `Summary` and find the `mean()` across the humidities.

Set the `as_index` parameter to `True` in order to have the indices of each mean humidity be the associated summary.
Set the `as_index` parameter to `False` in order to have the indices of each summary and associated mean humidity to be numerically ordered, returning a dataframe with both columns accessible.


```
summary_humidity_df = df.loc[:, ["Summary", "Humidity"]]
display(summary_humidity_df.groupby(["Summary"],as_index=False).mean())
```

# Joining/Merging Data (OPTIONAL)

Sometimes, your data may come from different sources, or you will want to add data to your dataset. Pandas has built-in merging and joining functionality that you can take advantage of to do this.

For example, let's say you want to create a new column that includes all of the temperatures at Fahrenheit. You can create this column and merge it with the existing data, using the `join()` function of a dataframe.

There are many ways that these join operations can happen, so I highly suggest reading the documentation/user guide when you are looking to perform your own merging operations (https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html). 

Here, we are using the most simple version, where the indices of the data we are joining match up. Thus, we can simply call `original_dataframe.join(new_dataframe)`, which returns the merged dataframe, in order to merge the data.

Let's look at how to do this with the Celsius to Fahrenheit conversion mentioned before.

```
fahrenheit = (df.loc[:,["Temperature (C)"]] * 1.8) + 32
fahrenheit = fahrenheit.rename(columns={"Temperature (C)": "Temperature (F)"})

extended_df = df.join(fahrenheit)
display(extended_df)
```

Let's take a closer look at this code.

First, you create a dataframe with only the column you are looking to modify. You can see here that doing addition and multiplication with scalar values (single numbers) means that the operation is applied to every value in the dataframe. Thus, every Celsius temperature is converted to Fahrenheit.

We want to change the column name to reflect the unit change, so we use `dataframe.rename()` which takes in a mapping from old names to new names.

Then, finally, we run our join. The indices match already beacuse we used data with these indices to generate our new data. Thus, the join is very simple and clean.
